```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import data from github**

```python
url = 'https://raw.githubusercontent.com/Sabrina-Hendricks/DS4002-Group13/main/Data/Cleaned_Data.csv'
df = pd.read_csv(url)
df.head()
```

**Graph 1: Histogram of the number of reviews per product type**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Class Name', order=df['Class Name'].value_counts().index) #Select product type column
plt.xticks(rotation=90)
plt.title('Number of Reviews per Product Type')
plt.xlabel('Product Type')
plt.ylabel('Number of Reviews')
plt.show()
```

This graph tells us that there are most reviews for dresses.

**Graph 2: Age distribution of reviewers**

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=15, kde=True) #Select age column
plt.title('Distribution of Age of Reviewers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Sentiment score distribution**

```python
sns.histplot(df['Sentiment Score'], bins=15, kde=True)
plt.title('Distribution of Sentiment Scores')
```

This graph suggests to us that most reviews have a positive sentiment.

**Sentiment score by age group**

```python
sns.barplot(data=df, x='Sentiment Score', hue='Age Group')
plt.title('Average Sentiment Score by Age Group')
```
