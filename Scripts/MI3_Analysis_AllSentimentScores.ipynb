```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

**Import Cleaned Data from GitHub**

```python
url = 'https://raw.githubusercontent.com/Sabrina-Hendricks/DS4002-Group13/main/Data/Cleaned_Data.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.describe() #looking at composition of data
```

**Examine Data**

Correlation Matrix for All Variables

```python
# Select only numeric columns from the dataframe for correlation calculation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix for the numeric columns
correlation_matrix = numeric_df.corr()

# Plot a heatmap for visualization
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix for Numerical Variables in the Dataframe")
plt.show()
```

Based on the matrix, we will look further into the relationship between Review Length and Sentiment Score. We will also look into Age and Product Type as they were a part of our hypothesis.

Check correlation between Age and Sentiment Score

```python
correlation = df['Age'].corr(df['Sentiment Score'])
print("Correlation between Age and Sentiment Score: ", correlation)
```

Regress Sentiment Score on Age

```python
np.random.seed(102) # Set the seed for the random number generator
N = 30

x = 5 + 2*np.random.normal(0,1,size = N) # Create an x
eps = np.random.normal(0,3,size = N) # Create noise
b0 = -1 # Intercept coefficoent
b1 = 3 # Slope coefficient
y = b0 + b1*x + eps

def slr(x,y): # Single Linear Regression Function
    x_bar = np.mean(x)
    y_bar = np.mean(y)
    b1 = np.inner(x-x_bar,y-y_bar)/np.inner(x-x_bar,x)
    b0 = y_bar - b1*x_bar
    y_hat = b0 + b1*x
    residuals = y - y_hat
    return({'b0':b0,'b1':b1,'y_hat':y_hat,'residuals':residuals})
```

```python
x = df['Age']
y = df['Sentiment Score']

coef = slr(x,y) # Single Linear Regression print('Intercept: ',coef['b0'], '\n', ' Slope: ', coef['b1'])

y_hat = coef['b0']+coef['b1']*x # Compute predictions

#Scatter plot of fit:
plt.scatter(x,y,label='Data')
plt.plot(x,y_hat,label='Regression Line',color='black')
plt.xlabel("Age")
plt.ylabel("Sentiment Score")
plt.legend(loc='upper right')
plt.title('Linear Regression of Sentiment Score on Age')

plt.show()
```

Regress Sentiment Score on Review Length

```python
x = df['Review Length']
y = df['Sentiment Score']

coef = slr(x,y) # Single Linear Regression print('Intercept: ',coef['b0'], '\n', ' Slope: ', coef['b1'])

y_hat = coef['b0']+coef['b1']*x # Compute predictions

#Scatter plot of fit:
plt.scatter(x,y,label='Data')
plt.plot(x,y_hat,label='Regression Line',color='black')
plt.xlabel("Review Length")
plt.ylabel("Sentiment Score")
plt.legend(loc='upper right')
plt.title('Linear Regression of Sentiment Score on Review Length')

plt.show()
```

Exploring relation with Product Type - Bar plot

```python
df['Sentiment Label'] = df['Sentiment Score'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Grouping by 'Class Name' and 'Sentiment Label' to get counts
class_sentiment_counts = df.groupby(['Class Name', 'Sentiment Label']).size().unstack()

# Plotting the bar plot for product types (Class Name) and sentiment
class_sentiment_counts.plot(kind='bar', stacked=True, figsize=(12, 8))

# Adding labels and title
plt.title('Count of Positive and Negative Reviews by Product Type (Class Name)')
plt.xlabel('Product Type (Class Name)')
plt.ylabel('Count of Reviews')
plt.legend(title='Sentiment')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

Correlation between Sentiment Scores and Product Types

```python
# Compute correlation between one-hot encoded Class Name columns and Sentiment Score
correlation_matrix = df[list(df.filter(regex='Class Name_').columns) + ['Sentiment Score']].corr()

# Extract only the correlations with Sentiment Score
correlation_with_sentiment = correlation_matrix['Sentiment Score'].drop('Sentiment Score')

# Display the correlation values
print(correlation_with_sentiment.sort_values(ascending=False))
```

Relationship with Age Group - Bar plot

```python
df['Sentiment Label'] = df['Sentiment Score'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Grouping by 'Age Group' and 'Sentiment Label' to get counts
age_group_sentiment_counts = df.groupby(['Age Group', 'Sentiment Label']).size().unstack()

# Plotting the bar plot for age group and sentiment
age_group_sentiment_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding labels and title
plt.title('Count of Positive and Negative Reviews by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Reviews')
plt.legend(title='Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
